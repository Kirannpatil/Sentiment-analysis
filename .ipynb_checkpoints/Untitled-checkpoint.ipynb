```python
# Basic Preprocessing
# Remove tags
# lowercase
# remove stopwords
```

```python
import re
def remove_tags(raw_text):
    cleaned_text = re.sub(re.compile('<.*?>'), '', raw_text)# Find the pattern '<.*?>' (html) and replace it with '' (nothing).
    return cleaned_text
```

```python
df['review']=df['review'].apply(remove_tags)
```

```python
df['review'][1]
```

```python
df['review']=df['review'].str.lower()
df['review'][1]
```

```python
from nltk.corpus import stopwords

sw_set = set(stopwords.words('english'))  # faster lookup

def clean_text(text):
    tokens = text.split()  # simple tokenization
    filtered = [word for word in tokens if word not in sw_set]
    return " ".join(filtered)

df['review'] = df['review'].apply(clean_text)
```

```python
from nltk.tokenize import word_tokenize
"""We use NLTK tokenization instead of a regex-based tokenizer because NLTK applies language-specific rules 
and handles contractions (shortened forms of words) and punctuation more intelligently. 
This helps preserve negation (e.g., “don’t” → “do + n’t”), which is important for sentiment analysis. 
Simple regex tokenizers may split contractions incorrectly and lose sentiment cues unless carefully designed."""

def tokenize_text(text):
    # Tokenize using NLTK
    return word_tokenize(text)

# Apply to your DataFrame
#df['tokens'] = df['review'].apply(tokenize_text)   hashed caused below is faster
df['tokens'] = [word_tokenize(text) for text in df['review']]
```
